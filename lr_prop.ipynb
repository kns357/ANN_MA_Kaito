```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import numpy as np
import scipy
```

```python
def relevance(r_k, activations, weights, j, k_range, j_range):
    r_j = np.sum((activations[j] * weights[j][k])/np.sum(activations[j_ite] * weights[j_ite][k] for j_ite in j_range) for k in k_range) * r_k
    return r_j
```
